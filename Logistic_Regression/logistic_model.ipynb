```python
# Load saved data
import numpy as np

X_train = np.load('X_train.npy')
X_test = np.load('X_test.npy')
y_train = np.load('y_train.npy')
y_test = np.load('y_test.npy')
```

```python
# Train Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty='l2',        
    max_iter=1000
)

model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Train Logistic Regression - with Class Balanced
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty='l2',        
    max_iter=1000,
    class_weight='balanced'
)

model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_balanced = model.predict(X_test)
```

```python
# Evaluate model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred_balanced))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_balanced))
print("Classification Report:\n", classification_report(y_test, y_pred_balanced))
```

```python
# Applying class weighting significantly improved the model’s ability to detect bankrupt companies 
# (recall increased from 14% to 82%), although this resulted in a 
# decrease in overall accuracy and an increase in false positives. This demonstrates the trade-off 
# between sensitivity and overall performance in imbalanced datasets.
```

```python
# Apply SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Check balance
import pandas as pd

print(pd.Series(y_train_smote).value_counts())
```

```python
# NOW scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_smote_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train model
from sklearn.linear_model import LogisticRegression

model_smote = LogisticRegression(
    penalty='l2',
    max_iter=1000
)
model_smote.fit(X_train_smote_scaled, y_train_smote)
```

```python
# Predictions
y_pred_smote = model_smote.predict(X_test_scaled)
```

```python
# Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred_smote))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_smote))
print("Classification Report:\n", classification_report(y_test, y_pred_smote))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predict probabilities
y_probs_orig = model.predict_proba(X_test_scaled)[:, 1]       # original model
y_probs_smote = model_smote.predict_proba(X_test_scaled)[:, 1]  # SMOTE model

# Compute ROC curves
fpr_orig, tpr_orig, _ = roc_curve(y_test, y_probs_orig)
fpr_smote, tpr_smote, _ = roc_curve(y_test, y_probs_smote)

# Compute AUC
auc_orig = roc_auc_score(y_test, y_probs_orig)
auc_smote = roc_auc_score(y_test, y_probs_smote)

# Plot
plt.figure(figsize=(8,6))
plt.plot(fpr_orig, tpr_orig, label=f'Original (AUC = {auc_orig:.3f})')
plt.plot(fpr_smote, tpr_smote, label=f'SMOTE (AUC = {auc_smote:.3f})')
plt.plot([0,1], [0,1], 'k--')  # random guess line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities
y_probs_orig = model.predict_proba(X_test_scaled)[:, 1]       # original model
y_probs_smote = model_smote.predict_proba(X_test_scaled)[:, 1]  # SMOTE model

# Compute ROC curves
fpr_orig, tpr_orig, _ = roc_curve(y_test, y_probs_orig)
fpr_smote, tpr_smote, _ = roc_curve(y_test, y_probs_smote)

# Compute AUC
auc_orig = auc(fpr_orig, tpr_orig)
auc_smote = auc(fpr_smote, tpr_smote)

# Plot
plt.figure(figsize=(8,6))
plt.plot(fpr_orig, tpr_orig, label=f'Original Model (AUC = {auc_orig:.3f})')
plt.plot(fpr_smote, tpr_smote, label=f'SMOTE Model (AUC = {auc_smote:.3f})', linestyle='--')
plt.plot([0,1], [0,1], color='grey', linestyle=':')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```
